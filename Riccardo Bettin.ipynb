```python
filtro = df.loc[:,['continent', 'location', 'total_cases','date']]
df_continent = filtro.set_index('continent')
df_continent
```

```python
df_continent.fillna(0,inplace=True)
```

```python
continente = input("Nome del continente: ").title() 
if continente in df_continent.index:
    cont_df = df_continent.loc[continente]
    if 'date' in cont_df.columns:
        cont_df = cont_df.sort_values(by='date')
    last_cases = cont_df.groupby('location')['total_cases'].last()
    totale_continente = last_cases.sum()
    print(f"Totale casi registrati nel continente {continente}: {totale_continente}")
else:
    print("Continente non trovato nel DataFrame.")
```

```python
last_global = df_continent.groupby('location')['total_cases'].last()
total_global = last_global.sum()
print(f"Persone infette totali nel mondo: {total_global}")
```
